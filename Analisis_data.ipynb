```python
import pandas as pd
import os
import statistics
import matplotlib.pyplot as plt
from matplotlib.pyplot import hist

directorioSalida="Data/:mes:/Consolidado/"
meses=["Junio","Julio","Agosto","Septiembre","Octubre"]

for i in range(len(meses)):
    correlacion = []
    for folder_depto in os.listdir(directorioSalida.replace(":mes:",meses[i])):            
            for csvFile in os.listdir(directorioSalida.replace(":mes:",meses[i])+folder_depto):
                csv_consolidado = pd.read_csv(directorioSalida.replace(":mes:",meses[i])+folder_depto+"/"+csvFile,delimiter="\t")
                canciones_mas_sonadas_depto=csv_consolidado.drop_duplicates('Title', keep='last')
                print("-----------Canciones más populares en "+folder_depto+" en el mes de "+meses[i])
                #hist(canciones_mas_sonadas_depto.Title, weights=canciones_mas_sonadas_depto.popularidad)
                #plt.pause(0.05)
                canciones_mas_sonadas_depto = canciones_mas_sonadas_depto.sort_values(by=['popularidad'], ascending=False)[0:5]
                print(canciones_mas_sonadas_depto[['Title','Artist','Reproducciones','popularidad']])
                canciones_mas_sonadas_depto = canciones_mas_sonadas_depto[['Title','Artist','Reproducciones','popularidad']]
                canciones_mas_sonadas_depto.to_csv( directorioSalida.replace(":mes:",meses[i])+folder_depto+"/"+"top10.csv",sep="\t")
                #Revisamos correlacion entre cantidad de reproducciones en emisora y popularidad en spotify
                print("-------------------------------------------------")
                correlacion.append(canciones_mas_sonadas_depto['Reproducciones'].corr(canciones_mas_sonadas_depto['popularidad']))
    print("Correlación mes "+meses[i]+": ",statistics.mean(correlacion))
```
