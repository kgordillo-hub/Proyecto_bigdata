```python
import pandas as pd
import os
from pandas import ExcelWriter
from pandas import ExcelFile

data_emisora = pd.read_csv("Emisoras_filtradas.csv",delimiter="\t")
directorioBase="Data/:mes:/Datasets/"
directorioSalida="Data/:mes:/Curado_datos/"
meses=["Junio","Julio","Agosto","Septiembre","Octubre"]
for index, row in data_emisora.iterrows():
    id_emisora= row['Etiquetas de fila']
    nomreEmisora = row['Nombre Streaming']
    departamento = row['Departamento']
    print(id_emisora)
    for i in range(len(meses)):
        directorio = directorioBase.replace(":mes:",meses[i])
        directorio_salida=directorioSalida.replace(":mes:",meses[i])
        for file in os.listdir(directorio):
            if id_emisora in file:
                #print(file)
                data_emisora = pd.read_excel(directorio+file, sheet_name="music")
                canciones_mas_sonadas=data_emisora.groupby(['Title','Artist', 'Spotify']).size().reset_index(name="Reproducciones").sort_values(by=['Reproducciones'],ascending=False)[0:10]
                if not os.path.exists(directorio_salida+departamento):
                    os.makedirs(directorio_salida+departamento)
                canciones_mas_sonadas.to_csv(directorio_salida+departamento+"/Top10_emisora_"+nomreEmisora+"_"+departamento+".csv",sep="\t")
```
